```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-1, 1, 100)
y = abs(100 * x)  # Line through (0,0) and (1,100)

plt.figure(figsize=(8, 6))
plt.plot(x, y, color='steelblue', linewidth=2)
plt.xlim(-1, 1)
plt.ylim(0, 100)
plt.xlabel('distance from center, %')
plt.ylabel('centering force, pixels per tick')
plt.title('Linear centering curve')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.savefig('plot.png', dpi=150)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-1, 1, 100)
y = abs(100 * x) + 5  # Line through (0,0) and (1,100)

plt.figure(figsize=(8, 6))
plt.plot(x, y, color='steelblue', linewidth=2)
plt.xlim(-1, 1)
plt.ylim(0, 105)
plt.xlabel('distance from center, %')
plt.ylabel('centering force, pixels per tick')
plt.title('Centering curve with linear and constant components')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.savefig('plot.png', dpi=150)
plt.show()
```
